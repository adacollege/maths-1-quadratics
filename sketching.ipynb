> Sketch the graph of $$y=3x^2+4x-1$$
> showing clearly the turning point and any points of intersection with the axes

### On the $y$-axis

$x=0$ so $y=-1$ and the curve passes through $(0,-1)$

### On the $x$-axis

$y=0$ so $3x^2+4x-1=0$

The discriminant is $4^2-4\cdot 3\cdot (-1)=28>0$ so this has two real roots.

Using the quadratic formula,
$$\begin{aligned}
x&=\dfrac{-b\pm\sqrt{b^2-4ac}}{2a}\\
&=\dfrac{-4\pm\sqrt{28}}{6}\\
&=\dfrac{-4\pm2\sqrt{7}}{6}\\
&=\dfrac{-2\pm\sqrt{7}}{3}
\end{aligned}$$

Hence the $x$-intercepts are at $x=\dfrac{-2-\sqrt{7}}{3}$ and $x=\dfrac{-2+\sqrt{7}}{3}$

### For the turning point

Completing the square

$$\begin{aligned}
3x^2+4x-1&=3(x+P)^2+Q\\
&=3(x^2+2Px+P^2)+Q\\
3x^2+4x-1&=3x^2+6Px+3P^2+Q
\end{aligned}$$

$$3x^2+4x-1=3x^2+6Px+3P^2+Q$$

Equating $x$-coefficients

$$4=6P\implies P=\frac{2}{3}$$

$$3x^2+4x-1=3x^2+6Px+3P^2+Q$$
Equating constants
$$\begin{aligned}
-1&=3P^2+Q\\
&=3\left(\frac{2}{3}\right)^2+Q\\
\implies Q&=-1-\frac{4}{3}\\
&=-2\frac{1}{3}
\end{aligned}$$

So $$y=3\left(x+\frac{2}{3}\right)^2-2\frac{1}{3}$$

And the turning point is at $\left(-\frac{2}{3},-2\frac{1}{3}\right)$

```python
import toyplot
import numpy
```

```python
def f(x): return 3*x**2+4*x-1
```

```python
xs = numpy.linspace(-2,1,100)
ys = f(xs)
canvas, axes, marks = toyplot.plot(xs,ys,width=400)
axes.x.spine.position = 0
axes.y.spine.position = 0
axes.x.ticks.show = False
axes.y.ticks.show = False
axes.scatterplot([-2/3,0,(-2-7**0.5)/3,(-2+7**0.5)/3],[f(-2/3),f(0),0,0],color=["green","red","blue","blue"],size=5);
axes.text(-2/3,f(-2/3)-0.3,"(-2/3,-2 1/3)")
axes.text(0.2,-1,"-1")
axes.text(-1.5,0.2,"(-2-<span>&#8730;</span>7)/3")
axes.text(0.5,0.2,"(-2+<span>&#8730;</span>7)/3");
```
